```python
from Alexandre.Class.Scrapper import Scrapper

scheme={
            'div':{
                'header':{
                    'div':{'class_':'D(ib)'}
                },
                'class_':'D(tbr)',
                'row':{
                    'div':{'class_':'D(tbc)'}
                }
            }
        }

url='https://finance.yahoo.com/quote/EDF.PA/financials?p=EDF.PA'

Scrapper_Financials=Scrapper(scheme=scheme,header=header,type='table')
print(Scrapper_Financials(url=url,header=True))
```

```python
from Alexandre.Class.Scrapper import Scrapper

scheme={
    'tr':{
        'header':{
            'th':{
                'class_1':'C($tertiaryColor) Fz(xs) Ta(end)',
                'class_2':'Fw(400) Py(6px)'
            }
        },
        'class_':'BdT Bdc($seperatorColor) Ta(end) Fz(s) Whs(nw)',
        'row':{
            'td':{'class_':'Py(10px)'}
        }
    }
}

url=('https://uk.finance.yahoo.com/quote/EDF.PA/history?p=EDF.PA')

Scrapper_History=Scrapper(scheme=scheme,header=header,type='table')
print(Scrapper_History(url=url,header=False))
```

Let's store everything in an object called "Asset"...

Now let's scrap data from Reddit and do some preliminary sentiment analysis.

```python
from Alexandre.Class.Reddit import Reddit

Ukraine=Reddit(topic="Ukraine",
            top=500,
            attributes=['headlines','id','author','created_utc','score','upvote_ratio','url'])
France=Reddit(topic="France",
            top=500,
            attributes=['headlines','id','author','created_utc','score','upvote_ratio','url'])
Ukraine.topic_model_LDA()
France.topic_model_LDA()
Ukraine.sentiment_analysis()
France.sentiment_analysis()

print(Ukraine.redit_data)
print(France.redit_data)
```
